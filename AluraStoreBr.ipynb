### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

df_loja1 = pd.read_csv(url)
df_loja2 = pd.read_csv(url2)
df_loja3 = pd.read_csv(url3)
df_loja4 = pd.read_csv(url4)

df_loja1.head()
```

#1. Análise do faturamento


```python
faturamento_loja1 = df_loja1['Preço'].sum() + df_loja1['Frete'].sum()
faturamento_loja1_semfrete = df_loja1['Preço'].sum()
faturamento_loja2 = df_loja2['Preço'].sum() + df_loja2['Frete'].sum()
faturamento_loja2_semfrete = df_loja2['Preço'].sum()
faturamento_loja3 = df_loja3['Preço'].sum() + df_loja3['Frete'].sum()
faturamento_loja3_semfrete = df_loja3['Preço'].sum()
faturamento_loja4 = df_loja4['Preço'].sum() + df_loja4['Frete'].sum()
faturamento_loja4_semfrete = df_loja4['Preço'].sum()

print('Faturamento da Loja 1:' + str(faturamento_loja1))
print('Faturamento da Loja 1 sem frete:' + str(faturamento_loja1_semfrete))
print('Faturamento da Loja 2:' + str(faturamento_loja2))
print('Faturamento da Loja 2 sem frete:' + str(faturamento_loja2_semfrete))
print('Faturamento da Loja 3:' + str(faturamento_loja3))
print('Faturamento da Loja 3 sem frete:' + str(faturamento_loja3_semfrete))
print('Faturamento da Loja 4:' + str(faturamento_loja4))
print('Faturamento da Loja 4 sem frete:' + str(faturamento_loja4_semfrete))
```

```python


fig, fat_4lojas = plt.subplots()

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos_com_frete = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
faturamentos_sem_frete = [faturamento_loja1_semfrete, faturamento_loja2_semfrete, faturamento_loja3_semfrete, faturamento_loja4_semfrete]


fat_4lojas.bar(lojas, faturamentos_com_frete, label='Com frete')
bar_labels = ['Loja 1', 'Loja 2', 'Loja 3', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

fat_4lojas.bar(lojas, faturamentos_com_frete, label=bar_labels, color=bar_colors)

fat_4lojas.set_ylabel('valor faturado com frete')
fat_4lojas.set_title('Fatturamento das lojas do Sr Joao COM FRETE')

plt.show()
```

# 2. Vendas por Categoria


```python
def criar_grafico_pizza(df_loja, nome_loja):
    vendas_por_categoria = df_loja.groupby('Categoria do Produto')['Preço'].sum()
    labels = vendas_por_categoria.index

    plt.figure(figsize=(6,6))
    plt.pie(vendas_por_categoria, labels=vendas_por_categoria.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Vendas por Categoria na Loja {nome_loja}')
    plt.show()

criar_grafico_pizza(df_loja1, '1')
criar_grafico_pizza(df_loja2, '2')
criar_grafico_pizza(df_loja3, '3')
criar_grafico_pizza(df_loja4, '4')

```

# 3. Média de Avaliação das Lojas

```python
import pandas as pd

lojas = {
    'Loja 1': df_loja1,
    'Loja 2': df_loja2,
    'Loja 3': df_loja3,
    'Loja 4': df_loja4
}

medias_avaliacao = {}

for nome_loja, df_loja in lojas.items():
  media = df_loja['Avaliação da compra'].mean()
  medias_avaliacao[nome_loja] = media

print(medias_avaliacao)
```

```python
plt.bar(medias_avaliacao.keys(), medias_avaliacao.values(), color='skyblue')
plt.ylabel('Média das Avaliações das compras')
plt.title('Comparação das Médias de Avaliações das compras por Loja')
plt.ylim(3.5, 4.2)  # aumentar pra ver melhor
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

# 4. Produtos Mais e Menos Vendidos

```python
import matplotlib.pyplot as plt
import pandas as pd

# Dicionário com os DataFrames de cada loja
lojas = {
    'Loja 1': df_loja1,
    'Loja 2': df_loja2,
    'Loja 3': df_loja3,
    'Loja 4': df_loja4
}

# Loop para criar os gráficos de pizza
for nome_loja, df_loja in lojas.items():
    # Calcula os produtos mais vendidos da loja
    mais_vendidos = df_loja.groupby('Produto')['Preço'].sum().sort_values(ascending=False)
    top_3_mais_vendidos = mais_vendidos.head(3)
    print(top_3_mais_vendidos)

    # Cria o gráfico de pizza
    plt.figure(figsize=(6, 6))
    plt.pie(top_3_mais_vendidos, labels=top_3_mais_vendidos.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Top 3 Produtos Mais Vendidos - {nome_loja}')
    plt.show()
```

# 5. Frete Médio por Loja


```python

lojas = {
    'Loja 1': df_loja1,
    'Loja 2': df_loja2,
    'Loja 3': df_loja3,
    'Loja 4': df_loja4
}

medias_frete = {}

for nome_loja, df_loja in lojas.items():
  media = df_loja['Frete'].mean()
  medias_frete[nome_loja] = media

print(medias_frete)
```

```python
plt.bar(medias_frete.keys(), medias_frete.values(), color='skyblue')
plt.ylabel('Média das Avaliações das compras')
plt.title('Comparação das Médias de Avaliações das compras por Loja')
plt.ylim(10, 40)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```
